```python
from sys import stdout
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.signal import savgol_filter

from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def optimise_pls_cv(X, y, n_comp):
    # Define PLS object
    pls = PLSRegression(n_components=n_comp)

    # Cross-validation
    y_cv = cross_val_predict(pls, X, y, cv=10)

    # Calculate scores
    r2 = r2_score(y, y_cv)
    mse = mean_squared_error(y, y_cv)
    rpd = y.std()/np.sqrt(mse)
    
    return (y_cv, r2, mse, rpd)
```

```python
chem = "iron"
X = np.array(df)
y = wetchem[chem].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

r2s = []
mses = []
rpds = []
xticks = np.arange(1, 41)
for n_comp in xticks:
    y_cv, r2, mse, rpd = optimise_pls_cv(X, y, n_comp)
    r2s.append(r2)
    mses.append(mse)
    rpds.append(rpd)
    
        
        
    
```

```python
def plot_metrics(vals, ylabel, objective):
    with plt.style.context('ggplot'):
        plt.plot(xticks, np.array(vals), '-v', color='blue', mfc='blue')
        if objective=='min':
            idx = np.argmin(vals)
        else:
            idx = np.argmax(vals)
        plt.plot(xticks[idx], np.array(vals)[idx], 'P', ms=10, mfc='red')

        plt.xlabel('Number of PLS components')
        plt.xticks = xticks
        plt.ylabel(ylabel)
        plt.title('PLS')

    plt.show()
```

```python
plot_metrics(mses, 'MSE', 'min')
```

```python
plot_metrics(rpds, 'RPD', 'max')
```

```python
plot_metrics(r2s, 'R2', 'max')
```

```python
for chem in wetchem.columns:
    y = wetchem[chem]
    X = np.array(df)
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
    y_cv, r2, mse, rpd = optimise_pls_cv(X, y, 99)
    y_cv = [i[0] for i in y_cv]
    plt.figure(figsize=(6, 6))
    with plt.style.context('ggplot'):
        plt.scatter(y, y_cv, color='red')
        plt.plot(y, y, '-g', label='Expected regression line')
        z = np.polyfit(y, y_cv, 1)
        plt.plot(np.polyval(z, y), y, color='blue', label='Predicted regression line')
        plt.xlabel('Actual')
        plt.ylabel('Predicted')
        plt.title(chem)
        plt.legend()
        plt.plot()
        plt.savefig(f"scatterplots/pls_{chem}.png")
```
